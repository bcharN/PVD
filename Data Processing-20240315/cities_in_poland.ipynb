# grouping

```python
cip.loc[:,['City','Province']].groupby('Province').count()
```

```python
cip.loc[:,['Population','Province']].groupby('Province').sum()
```

```python
cip.query("Province == 'opolskie'").loc[:,['Population','Region']].groupby('Region').sum()
```

```python
cip.query("Province == 'opolskie'").loc[:,['Population','Region']].groupby('Region').sum().sort_values('Population', ascending=False)
```
