```python
import pandas as pd
```

```python
airl = pd.read_csv('krk-airlines.csv')
airl
```

```python
fli = pd.read_csv('krk-flights.csv') 
fli
```

```python
fpas = pd.read_csv('krk-passengers.csv') 
fpas
```

```python
fpas_fli = pd.merge(fpas,fli,how="outer")
fpas_fli
```

```python
fpas_fli_airl = pd.merge(fpas_fli, airl,how='outer')
ffa = fpas_fli_airl
ffa
```

    * list including flight number and destination (two columns)


```python
ffa.loc[:,['flight','to']].drop_duplicates()
```

    * flight list with the full name of the airline and the name of the aircraft


```python
ffa.loc[:,['airline','airplane']].drop_duplicates()
```

    * a list of passengers on a flight to London sorted by surname


```python
ffa.loc[:,['name','surname','to']].query("to == 'London'").sort_values('surname')
```

    * a list of women flying to Paris ordered by surname


```python
ffa.loc[:,['name','surname','to','gender']].query("gender == 'Female' and to == 'Paris'").sort_values('surname')
```

    * number of passengers on the flight to Berlin


```python
ffa.query("to == 'Berlin'").count()
```

    * number of men flying from Krakow


```python
ffa.query("gender == 'Male' and `from` == 'Krakow'").count()
```

    * number of passengers for each flight


```python
ffa.loc[:,['flight','name']].groupby("flight").count()
```
